```python
import uproot
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = uproot.open("/tf/Higgs-Boson-LHC-Collision-Detector/sigfcc_350.root")
tree = file["Delphes"]
tree["Particle"].show()
```

```python
pid = abs(tree["Particle/Particle.PID"].array(library = "np"))
pt = tree["Particle/Particle.PT"].array(library = "np")
eta = tree["Particle/Particle.Eta"].array(library = "np")
phi = tree["Particle/Particle.Phi"].array(library = "np")
charge = tree["Particle/Particle.Charge"].array(library = "np")

events = len(pt)
```

```python
d = []
u = []
s = []
c = []
b = []
t = []
prim_b = []
prim_t = []

for i in range(events):
    
    for j in range(len(pid[i])):
        if pid[i][j] == 1:
            d.append([i,j])
        elif pid[i][j] == 2:
            u.append([i,j])
        elif pid[i][j] == 3: 
            s.append([i,j])
        elif pid[i][j] == 4:
            c.append([i,j])
        elif pid[i][j] == 5:
            b.append([i,j])
        elif pid[i][j] == 6:
            t.append([i,j])
        elif pid[i][j] == 7:
            prim_b.append([i,j])
        elif pid[i][j] == 8:
            prim_t.append([i,j])
```

```python
pt_d = []
eta_d = []
phi_d = []

pt_u = []
eta_u = []
phi_u = []

pt_s = []
eta_s = []
phi_s = []

pt_c = []
eta_c = []
phi_c = []

pt_b = []
eta_b = []
phi_b = []

pt_t = []
eta_t = []
phi_t = []

pt_prim_d = []
eta_prim_d = []
phi_prim_d = []

pt_prim_t = []
eta_prim_t = []
phi_prim_t = []

for i in range(len(d)):
    
    pt_d.append(pt[d[i][0]][d[i][1]])
    eta_d.append(eta[d[i][0]][d[i][1]])
    phi_d.append(phi[d[i][0]][d[i][1]])

for i in range(len(u)):
    
    pt_u.append(pt[u[i][0]][u[i][1]])
    eta_u.append(eta[u[i][0]][u[i][1]])
    phi_u.append(phi[u[i][0]][u[i][1]])

for i in range(len(s)):
    
    pt_s.append(pt[s[i][0]][s[i][1]])
    eta_s.append(eta[s[i][0]][s[i][1]])
    phi_s.append(phi[s[i][0]][s[i][1]])

for i in range(len(c)):
    
    pt_c.append(pt[c[i][0]][c[i][1]])
    eta_c.append(eta[c[i][0]][c[i][1]])
    phi_c.append(phi[c[i][0]][c[i][1]])

for i in range(len(b)):
    
    pt_b.append(pt[b[i][0]][b[i][1]])
    eta_b.append(eta[b[i][0]][b[i][1]])
    phi_b.append(phi[b[i][0]][b[i][1]])

for i in range(len(t)):
    
    pt_t.append(pt[t[i][0]][t[i][1]])
    eta_t.append(eta[t[i][0]][t[i][1]])
    phi_t.append(phi[t[i][0]][t[i][1]])

for i in range(len(prim_b)):
    
    pt_prim_b.append(pt[prim_b[i][0]][prim_b[i][1]])
    eta_prim_b.append(eta[prim_b[i][0]][prim_b[i][1]])
    phi_prim_b.append(phi[prim_b[i][0]][prim_b[i][1]])

for i in range(len(prim_t)):
    
    pt_prim_t.append(pt[prim_t[i][0]][prim_t[i][1]])
    eta_prim_t.append(eta[prim_t[i][0]][prim_t[i][1]])
    phi_prim_t.append(phi[prim_t[i][0]][prim_t[i][1]])
```

```python
data_d = {
    "pt_d": pt_d,
    "eta_d": eta_d,
    "phi_d": phi_d,    
}

data_u = {
    "pt_u": pt_u,
    "eta_u": eta_u,
    "phi_u": phi_u,    
}

data_s = {
    "pt_s": pt_s,
    "eta_s": eta_s,
    "phi_s": phi_s,    
}

data_c = {
    "pt_c": pt_c,
    "eta_c": eta_c,
    "phi_c": phi_c,    
}

data_b = {
    "pt_b": pt_b,
    "eta_b": eta_b,
    "phi_b": phi_b,    
}

data_t = {
    "pt_t": pt_t,
    "eta_t": eta_t,
    "phi_t": phi_t,    
}

data_prim_d = {
    "pt_prim_d": pt_prim_d,
    "eta_prim_d": eta_prim_d,
    "phi_prim_d": phi_prim_d,    
}

data_prim_t = {
    "pt_prim_t": pt_prim_t,
    "eta_prim_t": eta_prim_t,
    "phi_prim_t": phi_prim_t,    
}
```

```python
data_d = pd.DataFrame(data_d)
data_u = pd.DataFrame(data_u)
data_s = pd.DataFrame(data_s)
data_c = pd.DataFrame(data_c)
data_b = pd.DataFrame(data_b)
data_t = pd.DataFrame(data_t)
data_prim_d = pd.DataFrame(data_prim_d)
data_prim_t = pd.DataFrame(data_prim_t)
```
